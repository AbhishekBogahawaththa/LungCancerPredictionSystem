```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("Lung Cancer.csv")
```

```python
num_features = ["age", "cholesterol_level", "bmi"]
data = df[num_features]
```

```python
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=num_features)
```

```python
data.hist(bins=20, figsize=(12, 4))
plt.suptitle("Distributions BEFORE MinMax Scaling", fontsize=14)
plt.show()
```

```python
data_minmax.hist(bins=20, figsize=(12, 4))
plt.suptitle("Distributions AFTER MinMax Scaling", fontsize=14)
plt.show()
```

```python
for col in num_features:
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    axes[0].hist(data[col], bins=20, color="skyblue")
    axes[0].set_title(f"{col} BEFORE Scaling")

    axes[1].hist(data_minmax[col], bins=20, color="lightgreen")
    axes[1].set_title(f"{col} AFTER Scaling")

    plt.show()
```
